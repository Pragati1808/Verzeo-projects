```python
df2=df.copy()
df3=pd.get_dummies(df2)
df3.head()
```

```python
df3=df3[['Profit','R&D Spend','Administration','Marketing Spend','State_California','State_Florida','State_New York']]
df3
```

```python
x = df3.iloc[:, 1:].values
y = df3.iloc[:,0].values
```

```python
x = x[:, :-1]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train,y_test = tts(x,y,test_size =0.4)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
df4 = pd.DataFrame(data=y_test,columns=['y_test'])
df4['y_predict']= y_pred
df4
```

```python
from sklearn import metrics
rsq = metrics.r2_score(y_test,y_pred)
print("r2 score = ",rsq)
```

```python
figure(figsize=(14, 6), dpi=80)
sns.lineplot(data=df4)
```

the given values clearly suggest how close are our prediction results to the given data.
thus the model is good to go !!
